### 네이버 뉴스 제목 가져오기
* requests library import
* user-agent 헤더 설정 필요

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
```

```python
url = 'https://news.naver.com/'
req_header = {
    'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/95.0.4638.69 Safari/537.36'
}
res = requests.get(url, headers=req_header)
print(type(res))
print(res.status_code)
print(res.ok)
```

* 응답헤더들과 요청헤더값 확인하기

```python
# 응답헤더들
res.headers
```

```python
# 요청헤더들
res.request.headers
```

* response에 포함된 text 추출하기
* 파싱하기 위해서 BeautifulSoup 객체생성
* 특정 html 엘리먼트를 추출하기 위해서 BeautifulSoup객체의 select(css 선택자) 함수 사용

```python
if res.ok:
    # 소스보기에서 보여지는 text
    html = res.text
    soup = BeautifulSoup(html, 'html.parser')
    print(type(soup))
    print(len(soup.select("a[href*='read.naver']")))
    atags = soup.select("a[href*='read.naver']")
    print(type(atags))
    for atag in atags:
        print(type(atag), atag)
        print('-------------')
```
