```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import re

url = 'https://news.naver.com/'
req_header = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.114 Safari/537.36'
}
response = requests.get(url, headers=req_header)
print(response.status_code)
print(response.ok)

if response.ok:
    # response 객체에서 text 추출
    html = response.text
    # 추출한 텍스트를 파싱하기 위해서 BeautifulSoup 객체생성
    soup = BeautifulSoup(html, 'html.parser')
    
    news_list = []
    news_dict = {}
    
    for main in soup.select('div.main_component.droppable'):
        print(main.find('h4').text)
        if main.find('h4').text == '헤드라인 뉴스':
            #print(main.select('div.hdline_news ul li a'))
            atags = main.select("div.hdline_news ul li a[href*='read.naver']")
        else:    
            #print(main.select('div.com_list ul li a'))
            atags = main.select("div.com_list ul li a[href*='read.naver']")
            
        for atag in atags:
            #print(atag.text)
            news_dict['section'] = main.find('h4').text
            regex = re.compile(r'[\n]')
            result = regex.sub('',atag.text)
            news_dict['title'] = result
            news_dict['href'] = urljoin(url, atag['href'])
            #print(news_dict)    
            news_list.append(news_dict)
            news_dict = {}

news_list
```
